```python
import os
```

```python
LABEL_PATH="../../data_title/title/"
PAPER_PATH="../../data_title/abstract/"
```

경로내의 파일 리스트를 받아옵니다.

```python
label_list=os.listdir(LABEL_PATH)
paper_list=os.listdir(PAPER_PATH)
```

```python
label_dic={}
for label_el in label_list:
    label_dic[label_el]=1
```

```python
proc_dic={}
for paper_el in paper_list:
    if label_dic.get(paper_el, 0) == 1:
        proc_dic[paper_el]=1
    elif label_dic.get(paper_el, 0) == 0:
        proc_dic[paper_el]=10
    else:
        print("Special-case" + paper_el)
```

```python
import pandas as pd
```

파일을 읽어 1개의 문서로 만들기 위한 함수입니다.

```python
def file2df(filePath):
    ret=""
    with open(filePath, "r") as f:
        for line in f:
            ret = ret + line.replace("\n"," ")
        if len(ret) < 10:
            ret="No data"
    return ret
```

```python
col_names=["titles","abstracts"]
```

데이터가 있는 경로에서 파일을 읽어서 labels, papers 행으로 이루어진 DataFrame을 만듭니다.

```python
dfs = pd.DataFrame(columns=col_names)
for ky, vl in proc_dic.items():
    if (proc_dic[ky] == 1) and ("ids" not in ky):
        dfs=pd.concat([dfs, pd.DataFrame([[file2df(LABEL_PATH+"/"+ky),file2df(PAPER_PATH+"/"+ky)]], columns=col_names)], ignore_index=True)
```

```python
from gensim.models.word2vec import LineSentence
```

저장했었던 파일을 로드합니다.

```python
src_sentences_path="src_sentences.txt"
lines = LineSentence(src_sentences_path)
```

```python
from nltk.tag import pos_tag
from nltk.tag import untag
```

품사에 따라 불필요하다고 생각되는 품사들을 제외해서 만들어봅니다.

```python
tempL=[]
for line_el in lines:
    posl = pos_tag(line_el)
    newLine=[]
    for wd,tg in posl:
        if tg.startswith("NN"):
            newLine.append((wd,tg))
        if tg.startswith("VB"):
            newLine.append((wd,tg))
        if tg.startswith("JJ"):
            newLine.append((wd,tg))
    tempL.append(untag(newLine))
```

```python
tempL
```

품사제외 워드들의 숫자

```python
nWord=0
for tempEL in tempL:
    nWord = nWord + len(tempEL)
```

```python
nWord
```

기본 데이터의 워드들의 숫자

```python
srcWord=0
for line_el in lines:
    srcWord = srcWord + len(line_el)
```

```python
import os
```

파일에 백업합니다.

```python
src_sentences_path="shortences.txt"
with open(src_sentences_path, 'w',encoding="utf-8") as f:
    for tempEL in tempL:
        tempStr=""
        for one in tempEL:
            tempStr = tempStr+ " " + one
        f.write(tempStr+os.linesep)
```

딕셔너리로 만들어봅니다.

```python
wordmap={}
for tempEL in tempL:
    for wordEL in tempEL:
        wordmap[wordEL] = wordmap.get(wordEL,0) + 1
```

```python
wordmap
```

```python
import pandas as pd
```

딕셔너리를 DataFrame을 만들어봤습니다.

```python
s = pd.Series(wordmap, name='countings')
```

```python
s.index.name = 'words'
```

```python
wordmap_df = pd.DataFrame(s)
```

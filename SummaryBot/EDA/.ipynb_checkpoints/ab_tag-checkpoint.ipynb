# Import data and tagger
from nltk.corpus import twitter_samples
from nltk.tag import pos_tag_sents

# Load tokenized tweets
tweets_tokens = twitter_samples.tokenized('positive_tweets.json')

# Tag tagged tweets
tweets_tagged = pos_tag_sents(tweets_tokens)

# Set accumulators
JJ_count = 0
NN_count = 0

# Loop through list of tweets
for tweet in tweets_tagged:
    for pair in tweet:
        tag = pair[1]
        if tag == 'JJ':
            JJ_count += 1
        elif tag == 'NN':
            NN_count += 1

# Print total numbers for each adjectives and nouns
print('Total number of adjectives = ', JJ_count)
print('Total number of nouns = ', NN_count)

```python
from gensim.models.word2vec import LineSentence
```

```python
src_sentences_path="src_sentences.txt"
lines = LineSentence(src_sentences_path)
```

```python
isF=1
for line_el in lines:
    if isF==1:
        isF=0
        print(line_el)
```

```python
isF=1
for line_el in lines:
    if isF==1:
        isF=0
        posl = pos_tag(line_el)
        print(posl)
        newLine=[]
        for wd,tg in posl:
            if tg.startswith("NN"):
                newLine.append((wd,tg))
            if tg.startswith("VB"):
                newLine.append((wd,tg))
            if tg.startswith("JJ"):
                newLine.append((wd,tg))
        print(newLine)
        print(untag(newLine))
```

```python
tempL=[]
isF=100
for line_el in lines:
    if isF>0:
#         isF=isF-1
        posl = pos_tag(line_el)
#         print(posl)
        newLine=[]
        for wd,tg in posl:
            if tg.startswith("NN"):
                newLine.append((wd,tg))
            if tg.startswith("VB"):
                newLine.append((wd,tg))
            if tg.startswith("JJ"):
                newLine.append((wd,tg))
#         print(newLine)
#         print(untag(newLine))
        tempL.append(untag(newLine))
```
